```python
# Import the pyplot module from matplotlib as plt and make sure 
# plots appear in the notebook using '%matplotlib inline'
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# Create a simple plot using plt.plot()

# x axis values 
x = [1,2,3] 
# corresponding y axis values 
y = [2,4,1] 
    
# plotting the points  
plt.plot(x, y) 
    
# naming the x axis 
plt.xlabel('x - axis') 
# naming the y axis 
plt.ylabel('y - axis') 
    
    
# function to show the plot 
plt.show() 
```

```python
# Plot a single Python list

plt.plot([1, 2, 3, 4], [1, 4, 9, 16], 'ro')
plt.axis([0, 6, 0, 20])
plt.show()
```

```python
data = [3, 8, 2, 10, 5]

x = list(range(len(data)))

plt.plot(x, data, marker = '*', linestyle = '-', color = 'g', label = 'Data')

plt.xlabel('Index')
plt.ylabel('Value')
plt.title('Plot of a Single List')
plt.legend()
```

```python
# Create two lists, one called X, one called y, each with 5 numbers in them

x = [6, 7, 8, 9, 10]
y = [12, 14,16, 18, 20]

print("X :", x)
print("Y :", y)
```

```python
# Plot X & y (the lists you've created)
plt.plot(x, y, marker = '*', linestyle = '-', color = 'g', label = 'Data')

plt.xlabel('X')
plt.ylabel('Y')
plt.title('Plot of X & Y')
plt.legend()
```

There's another way to create plots with Matplotlib, it's known as the object-orientated (OO) method. Let's try it.



```python
fig, ax = plt.subplots()

ax.plot(x, y, marker = '*', linestyle = '-', color = 'g', label = 'Data')

ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_title('Plot of X & Y')
ax.legend()
```

```python
# Import and get matplotlib ready

import matplotlib.pyplot as plt

# Prepare data (create two lists of 5 numbers, X & y)

x = [1, 2, 3, 4, 5]
y = [8.5, 13, 5, 10.5, 10.5]

# Setup figure and axes using plt.subplots()

fig, ax = plt.subplots()

# Add data (X, y) to axes

ax.plot(x, y, marker = '^')

# Customize plot by adding a title, xlabel and ylabel

ax.set_xlabel('X - Axis')
ax.set_ylabel('Y - Axis')

ax.set_title('Custom Plot')

# Save the plot to file using fig.savefig()

fig.savefig('Custom_Plot.png')

plt.show()
```

```python
# Create an array of 100 evenly spaced numbers between 0 and 100 using NumPy and save it to variable X

x = np.linspace(0, 100, 100)

print(x)
```

```python
# Create a plot using plt.subplots() and plot X versus X^2 (X squared)

xsquared = x ** 2

fig, ax = plt.subplots()

ax.plot(x, xsquared, label = 'X^2')

ax.set_xlabel('X')
ax.set_ylabel('X^2')
ax.set_title('Plot of X versus X^2')
ax.legend()

plt.show()
```

```python
# Create a scatter plot of X versus the exponential of X (np.exp(X))

x = np.linspace(0, 5, 100)

exp_x = np.exp(x)

fig, ax = plt.subplots()

ax.scatter(x, exp_x, label = 'exp(X)', color = 'g', marker = '^')

ax.set_xlabel('X')
ax.set_ylabel('exp_x')
ax.set_title('Scatter Plot of X versus Exponential of X')
ax.legend()

plt.show()
```

```python
# Create a scatter plot of X versus np.sin(X)

x = np.linspace(0, 2 * np.pi, 100)

sin_x = np.sin(x)

fig, ax = plt.subplots()

ax.scatter(x, sin_x, label = 'Sin(X)', color = 'orange', marker = '*')

ax.set_xlabel('X')
ax.set_ylabel('Sin(X)')
ax.set_title('Scatter Plot of X versus np.sin(X)')
ax.legend()

plt.show()
```

```python
categories = ['Category A', 'Category B', 'Category C', 'Category D', 'Category E']
values = [25, 40, 30, 15, 50]

fig, ax = plt.subplots()

ax.bar(categories, values, color='purple')

ax.set_xlabel('Categories')
ax.set_ylabel('Values')
ax.set_title('Bar Plot Example')

plt.show()
```

```python
# Create a Python dictionary of 3 of your favourite foods with

favorite_foods = {'breakfast': 'Sigapparisi Idly',
                  'lunch': 'Panner Briyani',
                  'dinner': 'Poondu Dosai'}

for meal, food in favorite_foods.items():
    print(f"My favorite {meal} is {food}.")
    
print()
    
# The keys of the dictionary should be the food name and the values their price

food_prices = {'Sigapparisi Idly': 55,
               'Panner Briyani': 90,
               'Poondu Dosai': 35}

for food, price in food_prices.items():
    print(f"The price of {food} is {price:.2f}.")
```

```python
# Create a bar graph where the x-axis is the keys of the dictionary and the y-axis is the values of the dictionary

foods = list(food_prices.keys())
prices = list(food_prices.values())

plt.bar(foods, prices, color='pink')

# Add a title, xlabel and ylabel to the plot

plt.xlabel('Food')
plt.ylabel('Price ()')
plt.title('Food Prices')

plt.xticks(rotation=45, ha='right')

plt.tight_layout() #plt.tight_layout() to ensure that all labels and elements fit nicely within the figure
plt.show()
```
