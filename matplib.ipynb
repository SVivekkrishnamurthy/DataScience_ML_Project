## Object oriented ploting

```python
fig=plt.figure()
axes=fig.add_axes([0,0,1,1]) #Left, bottom, length, height
axes.plot(x,y1)
axes.set_xlabel('Data Scientist', fontsize=15)
axes.set_ylabel('Research Scientist', fontsize=15)
axes.set_title('job Role', fontsize=16)
```

```python
fig=plt.figure()
axes1=fig.add_axes([0,0,1,1]) #Left, bottom, length, height
axes2=fig.add_axes([0.12,0.6,0.5,0.3]) #Left, bottom, length, height
axes1.plot(x,y1,'v')
axes1.set_xlabel('Data Scientist', fontsize=15)
axes1.set_ylabel('Research Analyst', fontsize=15)
axes1.set_title('job Role', fontsize=16)
axes2.plot(x,y1,'b')
axes2.set_xlabel('Scientist', fontsize=15)
axes2.set_ylabel('Analyst', fontsize=15)
axes2.set_title('job Role', fontsize=16)
```
