How to apply custom defined functions?

```python
def my_mean(arr):
    return arr.sum()/len(arr)
```

```python
def divide_by_two(x):
    return x/2.0
```

```python
cyl=car.groupby('Cylinders')
```

```python
cyl.mean()
```

```python
cyl.agg(my_mean)
```

```python
car['Accer'].apply(divide_by_two)
```

```python
x=np.array([3,8,1,10])
plt.plot(x,'g',lw=10,marker='o',ms=15,mfc='orange')
plt.xlabel('Life')
plt.ylabel('Beautiful')
plt.title('Line Graph')
```

```python
x=np.array([3,8,1,10])
plt.plot(x,'g',lw=10,marker='o',ms=15,mfc='orange')
plt.xlabel('Life',{'family':'Verdana','color':'pink','size':18})
plt.ylabel('Beautiful',{'family':'Times New Roman','color':'blue','size':20})
plt.title('Line Graph',{'family':'Arial','color':'violet','size':16})
```

```python
#wave form creation usig Matplotlib
x=np.arange(0,5*np.pi,0.1)
y=np.sin(x)
plt.plot(x,y,'g',lw=8,marker='*',ms=14,mfc='orange')
plt.xlabel('Average Value')
plt.ylabel('Sine Value')
plt.title('Wave Form')
```

```python
#Bar chart
x=np.array([1,2,3,4])
y=np.array([15,18,100,60])
plt.bar(x,y)
plt.xlabel('Average Value')
plt.ylabel('Life Value')
plt.title('Span Value')
```

```python
#Bar chart
x=np.array([1,4])
y=np.array([15,60])
plt.bar(x,y,color='pink',width=0.5)
plt.xlabel('Average Value')
plt.ylabel('Life Value')
plt.title('Span Value')
```

```python
#Pie Chart
y=np.array([35,25,45,25])
fruits=['Apple','Orange','Banana','Kiwi']
plt.pie(y,labels=fruits)
```
