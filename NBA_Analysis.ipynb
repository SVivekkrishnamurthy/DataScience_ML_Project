```python
# creating bool series True for NaN values
bool_series = pd.isnull(data["Team"])
```

```python
# creating bool series False for NaN values
bool_series = pd.notnull(data["Salary"])
 
# displayed data only with team = NaN
data[bool_series]
```

```python
# isin() Method extracts rows from a DataFrame where a column value exists in a predefined collection
# creating a bool series from isin()
new = data["College"].isin(["Kansas"])
  
# displaying data with gender = male only
data[new]
```

```python
# creating filters of bool series from isin()
filter1 = data["College"].isin(["Kentucky"])
filter2 = data["Height"].isin(["6-6", "7-0", "6-10" ])
  
# displaying data with both filter applied and mandatory 
data[filter1 & filter2]
```

```python
# astype() - Method converts the data types in a Series
Emp_df4 = pd.read_csv("nba.csv")
# Printing the first 10 rows of 
# the data frame for visualization
  
Emp_df4[:10]
```

```python
# drop all those rows which 
# have any 'nan' value in it.
Emp_df4.dropna(inplace = True)
Emp_df4
```

```python
# let's find out the data type of Weight column
before = type(Emp_df4.Weight[0])
  
# Now we will convert it into 'int64' type.
Emp_df4.Weight = Emp_df4.Weight.astype('int64')
  
# let's find out the data type after casting
after = type(Emp_df4.Weight[0])
  
# print the value of before
print(before)
  
# print the value of after
print(after)
```

```python
 # Dataframe.sort_values() Single Parameter Sorting
    # sorting data frame by name
pt_csv.sort_values("Name", axis = 0, ascending = True,
                 inplace = True, na_position ='last')
 
# display
pt_csv
```

```python
# Changing position of Null values

# sorting data frame by name
pt_csv.sort_values("Salary", axis = 0, ascending = True,
                 inplace = True, na_position ='first')
 
pt_csv
# display
```

```python
# Dataframe.sort_values() Multiple Parameter Sorting:

#sorting data frame by Team and then By names
pt_csv.sort_values(["Team", "Name"], axis=0,
                 ascending=True, inplace=True)
  
#display
pt_csv
```

```python
#sorting data frame by Team and then By names
pt_csv.sort_values(["Team", "Name"], axis=0,
                 ascending=[True,False], inplace=True)
  
#display
pt_csv
```

```python
#sorting data frame by Team, age and height
pt_csv.sort_values(["Team", "Age", "Height"], axis=0,
                 ascending=[False,True,False],
inplace=True)
  
#display
pt_csv
```
